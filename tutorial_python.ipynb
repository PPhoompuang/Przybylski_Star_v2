```python
def plot(central, sem_range_x, range_y, wavelengths, array_of_data):
    plt.plot(wavelengths, array_of_data)
    plt.xlim(central-sem_range_x, central+sem_range_x)
    plt.ylim(0, range_y)
    plt.xlabel("Wavelength [Angstrom]")
    plt.ylabel("Intensity")
    plt.title(f"{central} Angstrom")
    plt.grid(True)
    return plt.show()
```
